```python
import ast              # Módulo de Árboles de Sintaxis Abstracta
import pandas as pd     # Librería para manipular datasets
import pyarrow as pa    # Útil para operaciones de lectura y escritura de datos
import pyarrow.parquet as pq   # Útil para leer y escribir datos en formato Parquet de manera eficiente
import os               # creación de directorios y comprobación de existencia
import matplotlib.pyplot as plt
from textblob import TextBlob
```

```python
# Cargar los archivos en DataFrames separados
df_user = pd.read_parquet('user_filtrado.parquet')
```

```python
df_review = pd.read_parquet('review_filtrado.parquet')
```

```python
# Fusionar los DataFrames en base al user_id
df_merged = pd.merge(df_review, df_user[['user_id', 'name']], on='user_id', how='left')
```

```python
df_merged
```

```python
columns_to_drop = ["useful", "funny", "cool"]
df_merged.drop(columns=columns_to_drop, inplace=True)
```

```python
df_merged
```

```python
# Cambiar el nombre de la columna 'stars' a 'rating' usando el método rename
df_Reviews = df_merged.rename(columns={'stars': 'rating'})
```

```python
df_Reviews
```

```python
# Número de filas por cada parte
tamano_parte = 1000000

# Calcular el número total de partes
num_partes = len(df_Reviews) // tamano_parte + 1

# Guardar cada parte en un archivo Parquet separado
for i in range(num_partes):
    inicio = i * tamano_parte
    fin = min((i + 1) * tamano_parte, len(df_Reviews))
    df_parte = df_Reviews.iloc[inicio:fin]
    nombre_archivo_parquet = f'Reviews_{i}.parquet'
    df_parte.to_parquet(nombre_archivo_parquet)
```
